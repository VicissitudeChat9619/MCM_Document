```python
def remove_chars_before_first_uppercase(s):
    for i, char in enumerate(s):
        if char.isupper():
   
            return s[i:]
    return ""
```

```python
data1=athletes_data[athletes_data['Year'] > 2000]

data1.index =range(len(data1))
data1.head()
```

```python
data2 = data1['NOC'].tolist()
NOCdata=list(set(data2))
print(NOCdata)
```

```python
data2 = data1['Sport'].tolist()
Sportdata=list(set(data2))
print(Sportdata)
```

```python
arr = np.zeros((len(Sportdata),len(NOCdata)))
sums = pd.DataFrame(arr,columns=NOCdata,index=Sportdata)
sums
```

```python
for i in range(len(data1)):
    if "No" in data1.loc[i,"Medal"]:
        continue
    sums.loc[data1.loc[i,"Sport"],data1.loc[i,"NOC"]]=sums.loc[data1.loc[i,"Sport"],data1.loc[i,"NOC"]]+1
sums
```

```python
# 创建二部图
import networkx as nx
B = nx.Graph()
B.add_nodes_from(NOCdata, bipartite=0)  # 添加国家节点
B.add_nodes_from(Sportdata, bipartite=1)    # 添加项目节点

# 添加边，边的权重为总奖牌数
for i in range(len(NOCdata)):
    for j in range(len(Sportdata)):
        B.add_edge(NOCdata[i], Sportdata[j], weight=sums.loc[Sportdata[j],NOCdata[i]])

# 检查二部图是否正确构建
print(nx.is_bipartite(B))
```

```python
B.degree()
```

```python
# 计算每个国家和项目的度
degree_dict = B.degree()

# 打印结果
for node, degree in degree_dict:
    print(f"{node}: {degree}")
```

```python
# 找出每个国家的核心项目
core_projects = {}
for country in NOCdata:
    country_edges = B.edges(country, data=True)
    if country_edges:
        max_weight_project = max(country_edges, key=lambda x: x[2]['weight'])[1]
        core_projects[country] = max_weight_project

# 打印核心项目
for country, project in core_projects.items():
    print(f"{country}: {project}")
```

```python
# 设置布局
pos = nx.spring_layout(B)

# 绘制节点
nx.draw_networkx_nodes(B, pos, nodelist=NOCdata, node_color='blue', label='Countries')
nx.draw_networkx_nodes(B, pos, nodelist=Sportdata, node_color='red', label='Sports')

# 绘制边
nx.draw_networkx_edges(B, pos, edge_color='gray')

# 添加标签
nx.draw_networkx_labels(B, pos, font_size=8)

# 显示图例
plt.legend()

# 显示图形
plt.show()
```
