```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data= pd.read_csv(
    '../2025_Problem_C_Data/summerOly_medal_counts.csv'
)
```

```python
data.head()
```

```python
year_g=data.groupby('Year')
years=list(set(data['Year']))
nocs=list(set(data['NOC']))
```

```python
noc_g=data.groupby('NOC')
noc_g.get_group('Canada')
```

```python
for i in years:
    i=int(i)
years=sorted(years)
print(years)
num=0
plt.figure()

for i in nocs:
    tem=noc_g.get_group(i)
    _x=tem['Year'].values.flatten()
    _y=tem['Total'].values.flatten()
    x=[]
    y=[]
    j=0
    for j in years:
        if j not in _x:

            x.append(j)
            y.append(0)
        else:
            x.append(j)
            y.append(_y[np.where(_x==j)[0][0]])
    if j==1: continue
    plt.plot(x,y,'-o')
    num+=1
plt.show()
print(num)
```

```python
x=[]
y=[]
for i in years:
    tem=year_g.get_group(i)
    x.append(i)
    y.append(tem['Total'].sum())
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('Total')
sum_total=y
```

```python
for i in nocs:
    tem=noc_g.get_group(i)
    _x=tem['Year'].values.flatten()
    _y=tem['Total'].values.flatten()
    x=[]
    y=[]
    j=0
    for j in years:
        if j not in _x:

            x.append(j)
            y.append(0)
        else:
            x.append(j)
            y.append(_y[np.where(_x==j)[0][0]]/sum_total[np.where(_x==j)[0][0]])
    if j==1: continue
    plt.plot(x,y,'-o')
    num+=1
```
